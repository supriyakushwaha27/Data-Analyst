```python
#Q4. Merge two dataframes on multiple columns(eg. city and Age)
df1=pd.DataFrame({
    'City':['Belapur','Ghansoli','Vashi','Turbhe','Thane'],
    'Age':[20,40,33,57,22],
    'Salary':[550000,222000,54600,40000,20000]
})
df2=pd.DataFrame({
    'City':['Belapur','Ghansoli','Turbhe','Kharghar','Lokhandwala'],
    'Age':[55,40,57,57,29],
    'Name':['Annie','Tom','Maddy','Hania','Sharren']
})
```

```python
merged_df=pd.merge(df1,df2,on=['City','Age'])
merged_df
```

```python
#Q5. Perform an inner join on two dataframes with a common column "Country"
df1=pd.DataFrame({
    'Country':['Qatar','India','Afghanistan','Bhutan','Nepal'],
    'Salary':[550000,222000,54600,40000,20000]
})
df2=pd.DataFrame({
    'Country':['India','Afghanistan','Bhutan','America','Greece'],
    'Name':['Annie','Tom','Maddy','Hania','Sharren']
})
```

```python
merged_df=pd.merge(df1,df2,on='Country',how='inner')
merged_df
```

```python
#Q6. Merge two dataframes and keep all the rows, even if there's no match in the second dataframe.
merged_df=pd.merge(df1,df2,on='Country',how='outer')
merged_df
```
